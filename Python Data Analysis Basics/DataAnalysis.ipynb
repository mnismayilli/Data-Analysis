As you can see the file contains many _NaN_ options. Let us work on adjusting them

```python
#Missing values dropna function in which axis = 0 -> for row and axis = 1-> for column
df.dropna(subset=["price"], axis=0, inplace=True)
```

```python
df.describe(include="all")
```

Some columns include _?_. So let us convert all _?_ to NaN.

```python
import numpy as np
df.replace("?", np.nan, inplace=True)
df.head()
```

Identify missing values

```python
#Identify missing value
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

Replace _NaN_ variables with the average mean

```python
#Calculate the average of the column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalised losses:", avg_norm_loss)
```

```python
#Replace missing values with the average of normalised losses
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
df["normalized-losses"].describe()
df["normalized-losses"].isnull()
```

```python
#Calculate the mean value for 'bore' column
avg_bore = df['bore'].astype('float').mean(axis=0)
print("Average of bore is ", avg_bore)
```

```python
df['bore'].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print('Average of stroke is', avg_stroke)
df['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

Replace _NaN_ observations with the most common type of the data in boolean variables

```python
#To see which values are present in a particular column
df['num-of-doors'].value_counts()
```

```python
#To see which values are the most common type in a particular column
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace=True)
```

```python
#Simply drop whole row with NaN in "price" column
df.dropna(subset=["price"],axis=0, inplace = True)
#reset index, because of dropped rows
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.dtypes
```

Let us change some variable types

```python
df[['bore', 'stroke']] = df[['bore', 'stroke']].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
df[['price']] = df[['price']].astype('float')
df[['peak-rpm']] = df[['peak-rpm']].astype('float')
```

```python
df.head()
```

## Data standardisation

```python
#Convert mpg to L/100 KM by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df['city-mpg']
#Check your transformed data
df.head()
```

```python
df['highway-L/100km'] = 235/df['highway-mpg']
```

```python
#Data Normalization
#replace originial value by original value/maximum value
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
df[['length', 'width', 'height']].head()
```

## Dummy variables

Let us get dummy variables for *fuel types*

```python
dummy_variable1 = pd.get_dummies(df['fuel-type'])
dummy_variable1.head()
```

Rename dummy variable columns

```python
dummy_variable1.rename(
    columns = {
        'gas': 'fuel-type-gas',
        'diesel': 'fuel-type-diesel'
    },
    inplace=True
)
dummy_variable1.head()
```

Let us now merge the new dummy variable to the data set

```python
df = pd.concat(
    [df, dummy_variable1],
    axis = 1
)
```

Drop the existing **fuel-type** from the data set

```python
#drop original column "fuel-type" from "df"
df.drop("fuel-type", axis=1, inplace=True)
df.head()
```

```python
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#Label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to center
ax.set_xticks(np.arange(grouped_pivot.shape[1])+0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0])+0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate lable if too long
plt.xticks(rotation =90)

fig.colorbar(im)
plt.show()
```

### Scipy library - some statistical tests

```python
from scipy import stats
```

```python
pearson_coef, p_value = stats.pearsonr(df['length'], df['price'])
print('The pearson correlation coefficient is ', pearson_coef, 'with a p-value of ', round(p_value,None))
```

p-value <0.0001 so the correlation between wheel-base and price is statistically significant but linear relationship is not extremely strong.
